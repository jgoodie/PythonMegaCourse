```python
import folium, pandas
volcanos = pandas.read_csv("../Volcanoes-USA.txt")
m = folium.Map(location=[45.372, -121.679], zoom_start=5, tiles='Stamen Terrain')
```

```python
for name,lat,lon in zip(volcanos['NAME'], volcanos['LAT'],volcanos['LON']):
    folium.Marker([lat, lon], popup=name, icon = folium.Icon(color = 'red')).add_to(m)
```
